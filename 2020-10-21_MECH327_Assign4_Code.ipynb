```python
def Cp_theta(T):
    
    C_0 = 1.10
    C_1 = -0.46
    C_2 = 1.0
    C_3 = -0.454
    
    theta = T / 1000 # [k]

    Cp_theta = C_0 + (C_1 * theta) + (C_2 * theta**2) + (C_3 * theta**3)
    
    return Cp_theta

def Cv_theta(T):
    
    k = 1.289
    
    C_0 = 1.10
    C_1 = -0.46
    C_2 = 1.0
    C_3 = -0.454
    
    theta = T / 1000 # [k]

    Cp_theta = C_0 + (C_1 * theta) + (C_2 * theta**2) + (C_3 * theta**3)
    Cv_theta = Cp_theta / k
    
    return Cv_theta
```
