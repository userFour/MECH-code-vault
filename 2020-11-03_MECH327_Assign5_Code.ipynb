```python
def Cp_theta(T):
    
    C_0 = 1.10
    C_1 = -0.46
    C_2 = 1.0
    C_3 = -0.454
    
    theta = T / 1000 # [k]

    Cp_theta = C_0 + (C_1 * theta) + (C_2 * theta**2) + (C_3 * theta**3)
    
    return Cp_theta

# ---------------------------------------------------------------------------

def Cv_theta5(T):
    
    C_0 = 1.05
    C_1 = -0.365
    C_2 = 0.85
    C_3 = -0.39
    
    R = 0.287
    
    theta = T / 1000 # [k]

    Cp_theta = C_0 + (C_1 * theta) + (C_2 * theta**2) + (C_3 * theta**3)
    Cv_theta = Cp_theta - R
    
    return Cv_theta

```

```python
# Question2

T0 = 300 # [K]
hfg = 2258.02
sfg = 6.0568

phi_steam = (-hfg) - T0*(-sfg)
print("phi_steam = %0.4f [kJ/kg]" % phi_steam)

# ----------------------------

h3 = 300.47
print("h3 = %0.4f [kJ/kg]" % h3)

xVals = np.array([320, 340])
yVals = np.array([320.58, 340.70])
h4 = np.interp(335, xVals, yVals)
print("h4 = %0.4f [kJ/kg]" % h4)

s0_3 = 6.86926
print("s0_3 = %0.4f [kJ/kgK]" % s0_3)

xVals = np.array([320, 340])
yVals = np.array([6.93413, 6.99515])
s0_4 = np.interp(335, xVals, yVals)
print("s0_4 = %0.4f [kJ/kg]" % s0_4)

phi_air = (h4 - h3) - T0*(s0_4 - s0_3)
print("phi_air = %0.4f [kJ/kg]" % phi_air)

q_air = h4 - h3
print("q_air = %0.4f [kJ/kg]" % q_air)

ratio = q_air / hfg
print("ratio = %0.4f []" % ratio)

phi_air_steam = phi_air * (1/ratio)
print("phi_air_steam = %0.4f [kJ/kg]" % phi_air_steam)

# ----------------------------

s_gen = (-1) * (1/T0) * (phi_steam + phi_air_steam)
print("s_gen = %0.4f [kJ/kgK]" % s_gen)

phi_destroyed = T0 * s_gen
print("phi_destroyed = %0.4f [kJ/kg]" % phi_destroyed)
```

```python
# Question3

T0 = 25 + 273.15
P0 = 100E3

T2 = 160 + 273.15
P2 = 800E3

mdot = 1
Qdot = 100E3
R = 0.287E3


h1 = 298.62
print("h1 = %0.4f [kJ/kg]" % h1)

xVals = np.array([420, 440])
yVals = np.array([421.59, 441.93])
h2 = np.interp(433.15, xVals, yVals)
print("h2 = %0.4f [kJ/kg]" % h2)


Wdot_act = (-Qdot/1E3) + mdot * (h1 - h2)
print("Wdot_act = %0.4f [kW]" % Wdot_act)


s0_1 = 6.86305
print("s0_1 = %0.4f [kJ/kgK]" % s0_1)

xVals = np.array([420, 440])
yVals = np.array([7.20875, 7.25607])
s0_2 = np.interp(433.15, xVals, yVals)
print("s0_2 = %0.4f [kJ/kgK]" % s0_2)

delta_s = s0_2 - s0_1 - (R/1E3)*log(P2/P0)
print("delta_s = %0.4f [kJ/kgK]" % delta_s)

s1 = s0_1
print("s1 = %0.4f [kJ/kgK]" % s1)
s2 = s0_1 + delta_s
print("s2 = %0.4f [kJ/kgK]" % s2)


s_gen = mdot*(s2 - s1) - (-Qdot/1E3)/T0
print("s_gen = %0.4f [kJ/kgK]" % s_gen)


Wdot_rev = T0*s_gen + Wdot_act
print("Wdot_rev = %0.4f [kW]" % Wdot_rev)


eta_2 = Wdot_rev / Wdot_act
print("eta_2 = %0.4f []" % eta_2)
```

```python
# Question4

T1 = 500 # [K]
P1 = 500 # [kPa]

P2 = 100 # [kPa]

R = 0.2968


# Set up the integration vars
n_steps = 10000
temperatures = np.linspace(T1, 1, n_steps)

delta_s = 0
residual = 1000
previousResidual = 1000

# IDEA: We can consider all the temperatures from T1 to
# 0 K. We can integrate delta_s for all these temperatures,
# and check the residual = delta_s - R*log(P2/P1) at each step.
# Every time the residual is smaller than the previous residual,
# we record its index. Eventually the residuals will begin increasing
# again as the sum delta_s - R*log(P2/P1) becomes negative, but we
# will still have the index of the smallest residual stored in memory

# Perform the integration with trapezoidal rule
for i, T in enumerate(temperatures[:-1]):

    # (b-a) * (1/2)*(f(a) + f(b))
    delta_s += (temperatures[i+1] - temperatures[i]) \
    * (1/2) * (Cp_theta(temperatures[i]) / temperatures[i] + \
              Cp_theta(temperatures[i+1]) / temperatures[i+1])
    
    residual = delta_s - R*log(P2/P1)
    
    if (abs(residual) < previousResidual):
        previousResidual = residual # update the recorded residual
        # print("Residual: %0.4f" % residual)
        bestIndex = i
    

print("Temperature for lowest residual: ", end="")
print(temperatures[bestIndex])


# -------------------- Verify --------------------

M = 28.01 # [kg/kmol]
s0_T1 = 212.833 # [kJ/kmol-K]
s0_1 = s0_T1 * (1/M)
print("\ns0_1 = %0.4f [kJ/kgK]" % s0_1)

s0_2 = s0_1 + R*log(P2/P1)
print("s0_2 = %0.4f [kJ/kgK]" % s0_2)

s0_T2 = s0_2 * M
print("s0_T2 = %0.4f [kJ/kgK]" % s0_T2)


xVals = np.array([197.831, 206.240])
yVals = np.array([300, 400])
T2 = np.interp(s0_T2, xVals, yVals)
print("T2 = %0.4f [K]" % T2)
```

```python
# Question5

T1 = 40 + 273.15 # [K]
P1 = 80 # [kPa]
CR = 4 # (v1/v2)


R = 0.287

# -------------------- Pt A --------------------
# Pv=RT
v1 = ((R*1E3) * T1) / (P1*1E3)
print("v1 = %0.4f [m^3/kg]" % v1)

# Set up the integration vars
n_steps = 30000
temperatures = np.linspace(T1, 545, n_steps)

delta_s = 0
residual = 1000
previousResidual = 1000

# Perform the integration with trapezoidal rule
for i, T in enumerate(temperatures[:-1]):

    # (b-a) * (1/2)*(f(a) + f(b))
    delta_s += (temperatures[i+1] - temperatures[i]) \
    * (1/2) * (Cv_theta5(temperatures[i]) / temperatures[i] + \
              Cv_theta5(temperatures[i+1]) / temperatures[i+1])
    
    residual = delta_s + R*log(1/CR)
    
    if (abs(residual) < abs(previousResidual)):
        previousResidual = residual # update the recorded residual
        # print("Residual: %0.4f" % residual)
        bestIndex = i
    
T2 = temperatures[bestIndex]
print("\nT2 lowest residual: %0.4f" % T2)

P2 = (R*T2) / (v1/4)
print("P2 = %0.4f [kPa]" % P2)


# -------------------- Pt B --------------------
# Recall first law --> du = Cv(T3 - T2) = dq

qin = 500

# Set up the integration vars
n_steps = 30000
temperatures = np.linspace(T2, 1150, n_steps)

delta_u = 0
residual = 1000
previousResidual = 1000

# Perform the integration with trapezoidal rule
for i, T in enumerate(temperatures[:-1]):

    # (b-a) * (1/2)*(f(a) + f(b))
    delta_u += (temperatures[i+1] - temperatures[i]) \
    * (1/2) * (Cv_theta5(temperatures[i]) + Cv_theta5(temperatures[i+1]))
    
    residual = delta_u - qin
    
    if (abs(residual) < abs(previousResidual)):
        previousResidual = residual # update the recorded residual
        # print("Residual: %0.4f" % residual)
        bestIndex = i
        
T3 = temperatures[bestIndex]
print("\nT3 lowest residual: %0.4f" % T3)

P3 = T3 * (P2/T2)
print("P3 = %0.4f [kPa]" % P3)


# -------------------- Pt C --------------------
# Recall that v2 = v3, v1 = v4

# Set up the integration vars
n_steps = 30000
temperatures = np.linspace(T3, 600, n_steps)

delta_s = 0
residual = 1000
previousResidual = 1000

# Perform the integration with trapezoidal rule
for i, T in enumerate(temperatures[:-1]):

    # (b-a) * (1/2)*(f(a) + f(b))
    delta_s += (temperatures[i+1] - temperatures[i]) \
    * (1/2) * (Cv_theta5(temperatures[i]) / temperatures[i] + \
              Cv_theta5(temperatures[i+1]) / temperatures[i+1])
    
    residual = delta_s + R*log(CR)
    
    if (abs(residual) < abs(previousResidual)):
        previousResidual = residual # update the recorded residual
        # print("Residual: %0.4f" % residual)
        bestIndex = i
    
T4 = temperatures[bestIndex]
print("\nT4 lowest residual: %0.4f" % T4)

P4 = (R*T4) / (v1)
print("P4 = %0.4f [kPa]" % P4)


# -------------------- Pt D --------------------
# Recall first law --> du = Cv(T4 - T1) = dq


# Set up the integration vars
n_steps = 1000
temperatures = np.linspace(T4, T1, n_steps)

delta_u = 0

# Perform the integration with trapezoidal rule
for i, T in enumerate(temperatures[:-1]):

    # (b-a) * (1/2)*(f(a) + f(b))
    delta_u += (temperatures[i+1] - temperatures[i]) \
    * (1/2) * (Cv_theta5(temperatures[i]) + Cv_theta5(temperatures[i+1]))
    
        
qout = delta_u
print("\nHeat rejected: %0.4f [kJ/kg]" % qout)


# -------------------- Pt E --------------------
# Recall first law efficiency = w_net / qin

# Set up the integration vars
n_steps = 1000
temperatures = np.linspace(T1, T2, n_steps)

delta_u = 0

# Perform the integration with trapezoidal rule
for i, T in enumerate(temperatures[:-1]):

    # (b-a) * (1/2)*(f(a) + f(b))
    delta_u += (temperatures[i+1] - temperatures[i]) \
    * (1/2) * (Cv_theta5(temperatures[i]) + Cv_theta5(temperatures[i+1]))
    
w12 = delta_u

# Set up the integration vars
n_steps = 1000
temperatures = np.linspace(T3, T4, n_steps)

delta_u = 0

# Perform the integration with trapezoidal rule
for i, T in enumerate(temperatures[:-1]):

    # (b-a) * (1/2)*(f(a) + f(b))
    delta_u += (temperatures[i+1] - temperatures[i]) \
    * (1/2) * (Cv_theta5(temperatures[i]) + Cv_theta5(temperatures[i+1]))
    
w34 = delta_u

w_net = abs(w12 + w34)
print("\nw_net = %0.4f [kJ/kg]" % w_net)

eta = w_net / qin
print("eta = %0.4f []" % eta)
    

# -------------------- Pt E --------------------
k = 1.4
eta_otto = 1 - CR**(1-k)
print("\neta_otto = %0.4f []" % eta_otto)
```
