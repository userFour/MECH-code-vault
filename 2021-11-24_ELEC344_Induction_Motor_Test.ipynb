```python
from math import *
import numpy as np
from scipy.optimize import fsolve
from scipy import signal
import matplotlib
import matplotlib.pyplot as plt

print("Sucess!")
```

```python
P = 4
kt = 0.5
s = 0.04
f = np.linspace(0, 10000, 1000000)
torque = np.zeros(len(f))

for i, w in enumerate(f):
    w_m = (1 - s) * f[i] * 2*pi
    torque[i] = kt * ((2*pi*f[i])/(P/2) - w_m)

plt.plot(f, torque)

plt.title("Torque-Speed Characteristics for a Motor")
plt.xlabel("Frequency [rad/s]")
plt.ylabel("Torque [Nm]")
#plt.ylim([0, 400])
#plt.xlim([0, 200])
plt.grid(True)
plt.legend()
plt.show()
```
