```python
# ---------------------- IMPORT THE LIBRARIES ----------------------
from math import *
import numpy as np
# from scipy.optimize import fsolve
import matplotlib
import matplotlib.pyplot as plt

print("Sucess!")
```

```python
np.set_printoptions(precision=4)

tVals = np.linspace(0, 1, int(1E3))
coords = np.zeros([2, len(tVals)])
#print(tVals)

# Compute the coordinates
for i, t in enumerate(tVals[:-1]):
    coords[0, i] = (0*t**3) + (1*t**2) + (0*t) + (-1)
    coords[1, i] = (1*t**3) + (0*t**2) + (-1*t) + (0)

#print(coords)

plt.plot(coords[0, :], coords[1, :])
plt.title("Parametric Curve")
plt.xlabel("x [ ]")
plt.ylabel("y [ ]")
plt.grid(True)
plt.show()
```
