```python
# Question2

P1 = 0.3E6 # [Pa]
T1 = 500 # [K]
P2 = 0.11E6 # [Pa]
T2 = 350 # [K]
Vdot = 1800 # [m^3/hr]

y_co2 =  0.19
M_co2 = 44.01
y_o2 = 0.0195
M_o2 = 31.999
y_n2 = 0.7905
M_n2 = 28.013

M_mix = y_co2 * M_co2 + y_o2 * M_o2 + y_n2 * M_n2
print("M_mix = %0.4f [kg/kMol]" % M_mix)

R_mix = R_u / M_mix
print("R_mix = %0.4f [kJ/kgK]" % R_mix)

Ndot = (P1 * Vdot) / (R_u * T1)
print("Ndot = %0.4f [Mol/hr]" % Ndot)
Ndot /= 3600
print("Ndot = %0.4f [Mol/s]" % Ndot)

delta_h_bar = y_co2 * (8305 - 2036) + y_o2 * (6086 - 1540.5) + y_n2 * (5911 - 1512.5)
print("delta_h_bar = %0.4f [kJ/kMol]" % delta_h_bar)


Wdot = (Ndot/1E3) * delta_h_bar # Convert Mol/s to kMol/s
print("Wdot = %0.4f [kJ/s]" % Wdot)
```

```python
# Question4

P1 = 1.05E5 # [Pa]
T1 = 20 + 273.15 # [K]
P2 = 2E5 # [Pa]
T2 = 100 + 273.15 # [K]
Vdot = 0.3 # [m^3/s]
R = 0.287E3 # [J/kgK]

phi_1 = 0.85

Pv1 = phi_1 * 2.339E3
print("Pv1 = %0.4f [kPa]" % (Pv1/1E3))

Pa1 = P1 - Pv1
print("Pa1 = %0.4f [kPa]" % (Pa1/1E3))

Pa2 = Pa1 * (T2/T1)
print("Pa2 = %0.4f [kPa]" % (Pa2/1E3))

Pv2 = P2 - Pa2
print("Pv2 = %0.4f [kPa]" % (Pv2/1E3))

Pg2 = 101.3E3 # [Pa]
phi_2 = Pv2/Pg2
print("phi_2 = %0.4f [ ]" % (phi_2))

# Compute compressor work

# Start by finding omega1, omega2
omega1 = (0.622*Pv1)/(P1 - Pv1)
# omega2 = (0.622*Pv2)/(P2 - Pv2)
omega2 = omega1
print("\nomega1 = %0.4f [ ]" % omega1)
print("omega2 = %0.4f [ ]" % omega2)


# Answer changes significantly based on whether hg comes from water, or ideal gas tables
# # Interpolate for hg1
# xVals = np.array([250, 300])
# yVals = np.array([461.36, 554.32])
# hg1 = np.interp(293.15, xVals, yVals)
# print("hg1 = %0.4f [kJ/kg]" % hg1)

# # Interpolate for hg2
# xVals = np.array([350, 400])
# yVals = np.array([647.90, 742.40])
# hg2 = np.interp(373.15, xVals, yVals)
# print("hg2 = %0.4f [kJ/kg]" % hg2)

# Get hg from water tables
hg1 = 2538.06 # [kJ/kg]
hg2 = 2676.05 # [kJ/kg]
print("hg1 = %0.4f [kJ/kg]" % hg1)
print("hg2 = %0.4f [kJ/kg]" % hg2)

# Interpolate for ha1
xVals = np.array([290, 298.15])
yVals = np.array([290.43, 298.62])
ha1 = np.interp(293.15, xVals, yVals)
print("ha1 = %0.4f [kJ/kg]" % ha1)

# Interpolate for ha2
xVals = np.array([360, 380])
yVals = np.array([360.86, 381.06])
ha2 = np.interp(373.15, xVals, yVals)
print("ha2 = %0.4f [kJ/kg]" % ha2)

# Compute h of mixture
h1_mix = ha1 + omega1 * hg1
h2_mix = ha2 + omega2 * hg2
print("h1_mix = %0.4f [kJ/kg_da]" % h1_mix)
print("h2_mix = %0.4f [kJ/kg_da]" % h2_mix)

Wdot = h2_mix - h1_mix
print("Wdot = %0.4f [kJ/kg_da]" % Wdot)

# Compute mass flow to get power
mdot = (Pa1 * Vdot) / (R * T1)
print("mdot = %0.4f [kg/s]" % mdot)
Power = Wdot * mdot
print("Power = %0.4f [kW]" % Power)
```

```python
# Question5

T1 = 25 + 273.15 # [K]
P1 = 101.325E3 # [Pa]
phi_1 = 0.65
T2 = 12 + 273.15 # [K]
P2 = P1
phi_2 = 1

Cp = 1.004

Pg1 = 3.169E3 # [kPa]

# Interpolate for Pg2
xVals = np.array([10, 15])
yVals = np.array([1.2276, 1.705])
Pg2 = np.interp(12, xVals, yVals)
Pg2 *= 1E3 # Convert to Pa
print("Pg2 = %0.4f [kPa]" % (Pg2/1E3))

# Compute Pv1, Pv2
Pv1 = phi_1 * Pg1
Pv2 = phi_2 * Pg2
print("Pv1 = %0.4f [kPa]" % (Pv1/1E3))
print("Pv2 = %0.4f [kPa]" % (Pv2/1E3))

# Compute relative humidities
omega_1 = (0.622 * Pv1) / (P1 - Pv1)
omega_2 = (0.622 * Pv2) / (P2 - Pv2)
print("omega_1 = %0.4f [kg_h20/kg_da]" % omega_1)
print("omega_2 = %0.4f [kg_h20/kg_da]" % omega_2)

# Compute mass flow condensate
m_dot_3 = omega_1 - omega_2
print("m_dot_3 = %0.4f [kg_h20/kg_da]" % m_dot_3)


# Find heat removed

# Interpolate for h3
xVals = np.array([10, 15])
yVals = np.array([41.99, 62.98])
h3 = np.interp(12, xVals, yVals)
print("\nh3 = %0.4f [kPa]" % h3)

hv1 = 2547.17
print("hv1 = %0.4f [kJ/kg]" % hv1)

# Interpolate for hv2
xVals = np.array([10, 15])
yVals = np.array([2519.74, 2528.91])
hv2 = np.interp(12, xVals, yVals)
print("hv2 = %0.4f [kJ/kg]" % hv2)

q_L = Cp*(T1 - T2) + (omega_1*hv1 - omega_2*hv2) - (omega_1 - omega_2)*h3
print("q_L = %0.4f [kJ/kg_da]" % q_L)
```
