```python
# Question 1

diameter1 = 48 # [in]
diameter2 = 48 # [in]

omega1 = 380 # [rpm]
omega2 = 380 # [rpm]

C = 16*12 # [in]

theta1 = pi #[]
theta2 = pi #[]

L = ((4*C**2) - (diameter1 - diameter2)**2)**(1/2) + (1/2)*(diameter1*theta1 + diameter2*theta2)
print("L = %0.4f [in]" % L)
L = L/12
print("L = %0.4f [ft]" % L)

velocity = (omega1*0.1047) * ((diameter1*25.4E-3)/2) # [m/s]
print("velocity = %0.4f [m/s]" % velocity)
velocity = velocity * 196.9 # [ft/min]
print("velocity = %0.4f [ft/min]" % velocity)

# Choose Polyamide A-4 belt
# Now we need to crunch formula to find forces
gamma = 0.039 # [lbf/in^3] specific weight
mu = 0.8 # [ ] coefficient of friction
t = 0.2 # [in] thickness of belt
# Guess b = width of belt = 4 in
b = 4 # [in]

w = 12 * gamma * b * t # [lbf/ft]
print("w = %0.4f [lbf/ft]" % w)

Fc = (w / 32.17) * (velocity / 60)**2
print("Fc = %0.4f [lbf]" % Fc)

# Find torque (equation on pg 888)
T = 63025 * 60 * 1.1 * 1 * (1/omega1) # [lbf*in]
print("T = %0.4f [lbf*in]" % T)

# Compute deltaF (equation on pg 885)
deltaF = 2*T / diameter1
print("deltaF = %0.4f [lbf]" % deltaF)

# Compute F1a
Cp = 1
Cv = 1
Fa = 175
F1a = b * Fa * Cp * Cv
print("F1a = %0.4f [lbf*in]" % F1a)

# Compute F2
F2 = -deltaF + F1a
print("F2 = %0.4f [lbf*in]" % F2)

# Compute Fi
Fi = (1/2)*(F1a + F2) - Fc
print("Fi = %0.4f [lbf*in]" % Fi)

# Compute transmitted horsepower
h_t = (1/33000) * (velocity * deltaF)
print("h_t = %0.4f [hp]" % h_t)

# Compute friction development
fprime = (1/pi) * log((F1a - Fc)/(F2 - Fc))
print("fprime = %0.4f [ ]" % fprime)

# Compute Fi using eqn 17-7
f = 0.8 # friction coefficient table 17-2
Fi = (T/diameter1) * (exp(f*theta1) + 1)/(exp(f*theta1) - 1)
print("Fi = %0.4f [lbf*in]" % Fi)

# Compute worst case dip using Fi
dip = ((C**2)*w)/(96*Fi)
print("dip = %0.4f [in]" % dip)
```
