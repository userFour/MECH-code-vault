```python
# ---------------------- IMPORT THE LIBRARIES ----------------------
from math import *
import numpy as np
# from scipy.optimize import fsolve
import matplotlib
import matplotlib.pyplot as plt

print("Sucess!")
```

```python
# ----- Q1 - Climb milling aluminium -----
#

dia = 3.0     #[in]
zn = 5       #[ ]
n = 16000     #[rpm]
doc_ax = 0.5  #[in]
doc_rad = 1.0 #[in]

# Compute cutting surface speed
Vc = (1/12) * (n*pi*dia)
print("Vc = \t%0.4f \t[ft/min]" % Vc)

# Verify in sensible units
# dia = dia*25.4E-3 [m]
# r = (dia*25.4E-3) / 2 [m]
# omega = 1675.5161 [rad/s]
# recall V = omega * r
print("Vc = \t%0.4f \t[m/s]" % (1675.5161 * ((dia*25.4E-3)/2)))

# Compute engagement angle
phi = degrees(acos(1 - ((2*doc_rad)/dia)))
print("phi = \t%0.4f \t[deg]" % phi)
fz = 0.004 / sin(radians(phi))
print("fz = \t%0.4f \t\t[in]" % fz)

# Compute programmed feedrate
Vf = fz * zn * n
print("Vf = \t%0.4f \t[in/min]" % Vf)

# Compute material removal rate
MRR = doc_rad * doc_ax * Vf
print("MRR = \t%0.4f \t[in^3/min]" % MRR)
MRR = MRR * 0.102 * 60
print("MRR = \t%0.4f \t[lb/hr]" % MRR)
```

```python
# ----- Q5 Bezier Curves -----
#

S = np.array([[0, 2, 5, 6], [0, 3, 2, 1]])
Q = np.array([[6.1, 7, 9, 10], [0.9, -1, -2, 1]])
resolution = int(1000)
uVals = np.linspace(0, 1, resolution)

SVals = np.zeros((2, resolution))
QVals = np.zeros((2, resolution))

# Plot the control points
plt.plot(S[0, :], S[1, :], 'bo', label="S Control")
plt.plot(Q[0, :], Q[1, :], 'ro', label="Q Control")

# Create curves
for i, u in enumerate(uVals):
  SVals[:, i] = (factorial(3) / (factorial(0) * factorial(3 - 0))) * (1 - u)**3 * S[:, 0] \
              + (factorial(3) / (factorial(1) * factorial(3 - 1))) * u**1 * (1 - u)**2 * S[:, 1] \
              + (factorial(3) / (factorial(2) * factorial(3 - 2))) * u**2 * (1 - u)**1 * S[:, 2] \
              + (factorial(3) / (factorial(3) * factorial(3 - 3))) * u**3 * S[:, 3]

  QVals[:, i] = (factorial(3) / (factorial(0) * factorial(3 - 0))) * (1 - u)**3 * Q[:, 0] \
              + (factorial(3) / (factorial(1) * factorial(3 - 1))) * u**1 * (1 - u)**2 * Q[:, 1] \
              + (factorial(3) / (factorial(2) * factorial(3 - 2))) * u**2 * (1 - u)**1 * Q[:, 2] \
              + (factorial(3) / (factorial(3) * factorial(3 - 3))) * u**3 * Q[:, 3]

plt.plot(SVals[0, :], SVals[1, :], 'b-', label="S Curve")
plt.plot(QVals[0, :], QVals[1, :], 'r-', label="Q Curve")

plt.title("Bezier Curves")
plt.xlabel("x [ ]")
plt.ylabel("y [ ]")
#plt.ylim([10, 150])
#plt.xlim([min(torque), max(torque)])
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
# ----- Q5 Bezier Curves - Superimpose Multiple Solutions -----
#

S2 = np.array([[5], [2]])
Q1 = np.array([[7], [-1]])
v = Q1 - S2
print(v)
numCurves = 3
colours = ['k-', 'c-', 'm-', 'b-', 'g-']
t = np.linspace(0.25, 0.75, numCurves)
joinPoints = np.zeros((2, numCurves))

for i in range(numCurves):
  joinPoints[:, i] = S2[:, 0] + t[i]*v[:, 0]

#print(joinPoints)

plt.plot(joinPoints[0, :], joinPoints[1, :], 'go', label="joinPoints")

S = np.array([[0, 2, 5, 6], [0, 3, 2, 1/2]])
Q = np.array([[6, 7, 9, 10], [1/2, -1, -2, 1]])
resolution = int(1000)
uVals = np.linspace(0, 1, resolution)

SVals = np.zeros((2, resolution))
QVals = np.zeros((2, resolution))

# Plot the control points
plt.plot(S[0, 0:3], S[1, 0:3], 'bo', label="S Control")
plt.plot(Q[0, 1:4], Q[1, 1:4], 'ro', label="Q Control")

# Create curves
for curveNum in range(numCurves):
  for i, u in enumerate(uVals):
    SVals[:, i] = (factorial(3) / (factorial(0) * factorial(3 - 0))) * (1 - u)**3 * S[:, 0] \
                + (factorial(3) / (factorial(1) * factorial(3 - 1))) * u**1 * (1 - u)**2 * S[:, 1] \
                + (factorial(3) / (factorial(2) * factorial(3 - 2))) * u**2 * (1 - u)**1 * S[:, 2] \
                + (factorial(3) / (factorial(3) * factorial(3 - 3))) * u**3 * joinPoints[:, curveNum]

    QVals[:, i] = (factorial(3) / (factorial(0) * factorial(3 - 0))) * (1 - u)**3 * joinPoints[:, curveNum] \
                + (factorial(3) / (factorial(1) * factorial(3 - 1))) * u**1 * (1 - u)**2 * Q[:, 1] \
                + (factorial(3) / (factorial(2) * factorial(3 - 2))) * u**2 * (1 - u)**1 * Q[:, 2] \
                + (factorial(3) / (factorial(3) * factorial(3 - 3))) * u**3 * Q[:, 3]

  plt.plot(SVals[0, :], SVals[1, :], colours[curveNum], label="t = {:.2f}".format(t[curveNum]))
  plt.plot(QVals[0, :], QVals[1, :], colours[curveNum])

plt.title("Bezier Curves - Multiple Solutions")
plt.xlabel("x [ ]")
plt.ylabel("y [ ]")
zoomScale = 1
# plt.ylim([min(joinPoints[1, :] - zoomScale), max(joinPoints[1, :] + zoomScale)])
# plt.xlim([min(joinPoints[0, :] - zoomScale), max(joinPoints[0, :]) + zoomScale])
plt.grid(True)
plt.legend(loc='best')
plt.show()
```
