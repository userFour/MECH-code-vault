```python
from math import *
import numpy as np
from scipy.optimize import fsolve
# import matplotlib
# import matplotlib.pyplot as plt

print("Sucess!")
```

```python
# Constants
gpm_to_m3 = 6.30901964E-5 # [gpm to m^3/s]
fps_to_mps = 0.3048 # [fps to m/s]
fpm_to_mps = 0.00508 # [fpm to m/s]
lb_to_kg = 0.45359237 # [lb to kg]
psi_to_pa = 6894.75729 # [psi to pa]
n_to_lbf = 0.224808943 # [n to lbf]
ft_to_m = 0.3048 # [ft to m]
kw_to_hp = 1.34102209 # [kW to hp]

print("ok")
```

```python
# Question1

P = 1500 # [psi]
dia_piston = 5 # [in]
dia_rod = 2 # [in]
GPM = 25

A_piston = (pi/4) * dia_piston**2
A_rodEnd = (pi/4) * (dia_piston**2 - dia_rod**2)
A_rod = (pi/4) * dia_rod**2


F_extension = P * A_rod
F_retraction = P * A_rodEnd
print("F_extension = %0.4f [lbf]" % F_extension)
print("F_retraction = %0.4f [lbf]" % F_retraction)


v_ext = 19.25*GPM / A_rod
v_ret = 19.25*GPM / (A_rodEnd)
print("v_ret = %0.4f [fpm]" % v_ret)
print("v_ext = %0.4f [fpm]" % v_ext)


# Recall that Q = v * A

# Compute hose diameter for inflow during extension
# First compute the various flows
Q_in = GPM
print("Q_in (extension or retraction) = %0.4f [GPM]" % Q_in)

Q_out_extension = (v_ext * A_rodEnd) / 19.25
print("Q_out_extension = %0.4f [GPM]" % Q_out_extension)

Q_out_retraction = GPM * (A_piston / A_rodEnd)
print("Q_out_retraction = %0.4f [GPM]" % Q_out_retraction)

# Compute hose for inflow
fxn = lambda x: ((Q_in * gpm_to_m3) / (pi/4 * x**2)) - (15 * fps_to_mps)
hose_in_dia = fsolve(fxn, 0.2)
print("hose_in_dia = %0.4f [m]" % hose_in_dia)
print("hose_in_dia = %0.4f [in]" % (hose_in_dia/25.4E-3))

# Compute hose for outflow (extension)
fxn = lambda x: ((Q_out_extension * gpm_to_m3) / (pi/4 * x**2)) - (15 * fps_to_mps)
hose_out_ret_dia = fsolve(fxn, 0.2)
print("hose_out_ext_dia = %0.4f [m]" % hose_out_ret_dia)
print("hose_out_ext_dia = %0.4f [in]" % (hose_out_ret_dia/25.4E-3))

# Compute hose for outflow (retraction)
fxn = lambda x: ((Q_out_retraction * gpm_to_m3) / (pi/4 * x**2)) - (15 * fps_to_mps)
hose_out_ret_dia = fsolve(fxn, 0.2)
print("hose_out_ret_dia = %0.4f [m]" % hose_out_ret_dia)
print("hose_out_ret_dia = %0.4f [in]" % (hose_out_ret_dia/25.4E-3))


# Note that maximum flow in the system will be after the motor and before the control valve
Q_max = Q_out_extension + 25
print("Q_max = Q_out_extension + 25 = %0.4f [GPM]" % Q_max)
hose_max_dia = ((4 * Q_max*gpm_to_m3) / (pi * 15*fps_to_mps))**(1/2)
print("hose_max_dia = %0.4f [m]" % hose_max_dia)
print("hose_max_dia = %0.4f [in]" % (hose_max_dia/25.4E-3))


```

```python
# Question3

P = 1000 * psi_to_pa # [Pa]
print("P = %0.4f [Pa]" % P)
mass = 6000 * lb_to_kg # [kg]
print("mass = %0.4f [kg]" % mass)

stroke_length = 20 * 25.4E-3 # [m]
mu = 0.15

g = 9.81

# Resolve the forces
A = np.array([[sin(radians(30)), sin(radians(60))], [cos(radians(30)), -cos(radians(60))]])
b = np.array([[mass*g], [0]])
x = np.linalg.solve(A, b)

F_p = x[0]
F_N = x[1]

print("F_p = %0.4f [N]" % F_p)
print("F_N = %0.4f [N]" % F_N)

# Also recall F_fr = mu * F_N
F_fr = mu * F_N
print("F_fr = %0.4f [N]" % F_fr)
F_p += F_fr
print("F_p = %0.4f [N]" % F_p)
print("F_p = %0.4f [lbf]" % (F_p*(1/lb_to_kg)))

# F = P * A
A = F_p / P
fxn = lambda x: ((pi/4) * x**2) - A
piston_dia = fsolve(fxn, 0.2)
print("piston_dia = %0.4f [m]" % piston_dia)
print("piston_dia = %0.4f [in]" % (piston_dia / 25.4E-3))


# ------------------------- Pt B -------------------------

v = 0.5 * fpm_to_mps # [m/s]
print("Final velocity = %0.4f [m/s]" % v)
d = 0.5 * 25.4E-3 # [m]
print("Distance = %0.4f [m]" % d)

# Recall vf^2 = vi^2 + 2ad but vi = 0
# a = vf^2 * (1/2d)
a = (v**2) * (1/(2*d))

acceleration_force = mass * a
print("acceleration_force = %0.4f [N]" % acceleration_force)
```

```python
# Question4

m = 2000 * lb_to_kg
P = 1000 * psi_to_pa
print("P = %0.4f [MPa]" % (P/1E6))
g = 9.81

F_p = ((7.5*ft_to_m)*(m*g)*cos(radians(45))) / ((2.5*ft_to_m)*sin(radians(30)))
print("F_p = %0.4f [N]" % F_p)
print("F_p = %0.4f [lbf]" % (F_p*n_to_lbf))

fxn = lambda x: P * (pi/4*x**2) - F_p
piston_dia = fsolve(fxn, 0.2)
print("piston_dia = %0.4f [m]" % piston_dia)
print("piston_dia = %0.4f [in]" % (piston_dia / 25.4E-3))

piston_dia = 3.5 * 25.4E-3

# ---------------------- Pt D ----------------------
# We want to lift load at 8in/s, what size motor/pump/hose do we need?
# Recall that hp = GPM x PSI x 0.000583
# To find GPM, we need to know cylinder extension rate
# or in metric, 1kW = LPM * bar/600
# From dynamics and stuff, we get Vcyl = 1.8856 in/s

Vcyl = 1.8856 * 25.4E-3 # [m/s]
print("\nVcyl = %0.4f [m/s]" % Vcyl)

piston_dia = 3.5 * 25.4E-3 # [3.5 in cylinder design choice]
# piston_dia = 100E-3 # [100mm cylinder design choice]
print("piston_dia = %0.4f [m]" % piston_dia)

A_piston = pi/4 * piston_dia**2

Q = A_piston * Vcyl
print("Q = %0.4f [m^3/s]" % Q)
print("Q = %0.4f [LPS]" % (Q*1E3))

power = (Q*1E3*60) * ((P/100E3) / 600)
print("power = %0.4f [kW]" % power)

# Recall that we have 15% power loss from motor to pump
power /= 0.85
print("Adjusted power = %0.4f [kW]" % power)
print("Adjusted power = %0.4f [hp]" % (power*kw_to_hp))


# Find the hose diameter
hose_dia = ((4 * Q) / (pi * 15*fps_to_mps))**(1/2)
print("hose_dia = %0.4f [mm]" % (hose_dia*1E3))
print("hose_dia = %0.4f [in]" % (hose_dia/25.4E-3))



# ---------------------- Pt E ----------------------
# Need to compute piston acceleration
v = Vcyl # [m/s]
print("\nFinal velocity = %0.4f [m/s]" % v)
d = 0.75 * 25.4E-3 # [m]
print("Distance = %0.4f [m]" % d)

# Recall vf^2 = vi^2 + 2ad but vi = 0
# a = vf^2 * (1/2d)
a = (v**2) * (1/(2*d))
print("Piston acceleration = %0.4f [m/s^2]" % a)

acceleration_load_perpendicular = 6*a
print("acceleration_load_perpendicular = %0.4f [m/s^2]" % acceleration_load_perpendicular)

acceleration_load = acceleration_load_perpendicular * cos(radians(45))
print("acceleration_load = %0.4f [m/s^2]" % acceleration_load)

F_accel = m * acceleration_load
print("F_accel = %0.4f [N]" % F_accel)

# Does it change the results?
F_p = ((7.5*ft_to_m)*(m*g + F_accel)*cos(radians(45))) / ((2.5*ft_to_m)*sin(radians(30)))
print("F_p = %0.4f [N]" % F_p)
print("F_p = %0.4f [lbf]" % (F_p*n_to_lbf))
```
