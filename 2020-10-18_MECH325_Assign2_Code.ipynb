```python
# Question2

p_nd = normal_diametral_pitch = 12 # [in]

phi_n = normal_pressure_angle = 20 # [deg]

N = num_teeth = 48 # []

face_width = 1.5 # [in]

# helix angle
psi = 45 # [deg]


# Compute circular pitch, axial pitch, pitch diameter, transverse pressure angle
# and number of axial pitches in face width

# Compute diametral pitch
# Diametral pitch is the ratio of
# the number of teeth in the gear to the pitch diameter.
p_d = p_nd * cos(radians(psi))
print("Diametral pitch, Pd = %0.4f [N/in]" % p_d)

# Compute transverse circular pitch
# The transverse circular
# pitch is the distance from a point on one tooth to the corresponding
# point on the next adjacent tooth, measured
# at the pitch line in the transverse plane.
p_t = pi / p_d
print("Transverse circular pitch, Pt = %0.4f [in]" % p_t)

# Compute normal circular pitch
# Normal circular pitch is
# the distance between corresponding points on adjacent
# teeth measured on the pitch surface in the normal direction.
p_n = p_t * cos(radians(psi))
print("Normal circular pitch = %0.4f [in]" % p_n)

# Compute axial pitch
# Axial pitch is the distance between
# corresponding points on adjacent teeth, measured on the
# pitch surface in the axial direction.
p_x = p_t / tan(radians(psi))
print("Axial pitch = %0.4f [in]" % p_x)
                     
# Compute pitch diameter
# Use eqn (8-15) Pg. 326
D = N / p_d
print("Pitch diameter D = %0.4f [in]" % D)

# Compute transverse pressure angle phi_t
# Use eqn (8-11) Pg. 324
phi_t = degrees(atan((tan(radians(phi_n)) / cos(radians(psi)))))
print("Transverse pressure angle phi_t = %0.4f [deg]" % phi_t)

# Compute the number of axial pitches in face width
num_axial_pitches = face_width / p_x
print("Number of axial pitches in face width = %0.4f []" % num_axial_pitches)


```

```python
# Question3

w_in = 2500 # [rpm]
w_out_min = 280 # [rpm]
w_out_max = 300 # [rpm]

pressure_angle = 20 # [deg]

transmitted_power = 25 # [hp]

# Determine suitable numbers of teeth for each gear while minimizing gearbox size
# and avoiding issues of interence

# First compute reduction ratio - recall that if we compute using w_out_max,
# we will achieve the most compact gearbox
desired_ratio = w_in / w_out_max
print("Desired ratio: %0.4f []" % desired_ratio)

# Let this be a 2 stage reduction --> ratio = (N1/N2)*(N3/N4)
# Compute the single stage ratio
m = desired_ratio ** (1/2)
print("m: %0.4f []" % m)

# Recall eqn (13-11) from Shigley 9e for smallest teeth on pinion
# Assume full depth teeth --> k = 1 Shigley 9e, Pg. 686
k = 1
N_p =  (2*k) / ((1+2*m)*(sin(radians(pressure_angle))**2)) \
* (m + ((m**2 + (1+2*m)*((sin(radians(pressure_angle))**2))))**(0.5))
print("Number of teeth on pinion: %0.4f []" % N_p)

# We need at least 15 teeth on pinon to avoid interference
# Compute the number of teeth on driven gear
N_driven = 15 * m
print("Number of teeth on driven gear: %0.4f []" % N_driven)

# Take number of teeth on driven gear 44
N_pinion = 15
N_driven = 44
w_out = w_in * (N_pinion / N_driven) * (N_pinion / N_driven)
print("Output RPM: %0.4f [rpm]" % w_out)
print("\n# ----------------------------------------------------------------------------- \n")

# 290 RPM is withing acceptable range

# -----------------------------------------------------------------------------

# Assume diametral pitch 6 teeth/in
# Determine:
# pitch diameters for each gear
# pitch line velocities in (ft/min)
# magnitudes of tangential, radial, and total forces transmitted between each set of gears
# input torque
# output torque, neglecting frictional losses

diametral_pitch = 6 # [teeth/in]
p_d = diametral_pitch

# Let pinion be 1, 3 driven gear be 2, 4

# Apply eqn (8-15) from Mott 63, Pg. 325
D1 = N_pinion / p_d
D2 = N_driven / p_d
print("Pinion diameter: %0.4f [in], Driven gear diameter: %0.4f [in]\n" % (D1, D2))


# Find pitch line velocities
# Pitch line velocity for first pinion and first driven gear will be the same
r1 = (1/2) * iTom(D1) # [m]
w1 = toAngularVelocity(w_in) # [rad/s]
v1 = w1 * r1 # [m/s]
v1 = v1 * 196.85039370079 # Convert to ft/min
print("Pinion1 pitch line velocity = Driven gear2 pitch line velocity = %0.4f [ft/min]" % v1)
# Pitch line velocity for the  second pinion and second driven gear
r2 = r1
w2 = toAngularVelocity(w_in * (N_pinion / N_driven))
v2 = w2 * r2 # [m/s]
v2 = v2 * 196.85039370079 # Convert to ft/min
print("Pinion3 pitch line velocity = Driven gear4 pitch line velocity = %0.4f [ft/min]\n" % v2)


# Find magnitude of forces
# Compute for input gears first by recalling eqn (13-35) Shigley 9e
print("For input gears:")
load_tangential = 33000 * (transmitted_power / v1)
print("Tangential force = %0.4f [lbf]" % load_tangential)
load_radial = load_tangential * tan(radians(pressure_angle))
print("Radial force = %0.4f [lbf]" % load_radial)
load_total = (load_tangential**2 + load_radial**2)**(0.5)
print("Total force = %0.4f [lbf]" % load_total)

print("For output gears:")
load_tangential = 33000 * (transmitted_power / v2)
print("Tangential force = %0.4f [lbf]" % load_tangential)
load_radial = load_tangential * tan(radians(pressure_angle))
print("Radial force = %0.4f [lbf]" % load_radial)
load_total = (load_tangential**2 + load_radial**2)**(0.5)
print("Total force = %0.4f [lbf]\n" % load_total)


# Compute the input torque
# Recall that torque = force * distance
load_tangential = 33000 * (transmitted_power / v1)
torque = load_tangential * (D1/2)
print("Input torque = %0.4f [lbf*in]" % torque)

# Compute the output torque
torque = torque * (44/15)**2
print("Output torque = %0.4f [lbf*in]" % torque)
```

```python
# Question7

Ng = 40
Nw = 1

w_out = 30 # [rpm]

pitch_diameter = D_worm = 1.25 # [in]
phi_n = 14.5 # [deg]
torque = 924 # [lb*in]

ratio = Nw/Ng


# Compute input rpm
w_in = w_out / ratio
print("w_in = %0.4f [rpm]" % w_in)


# Compute it using eqns from Shigley 9e, Pg. 817
p_x = pi / p_d
print("Axial pitch p_x = %0.4f [in]" % p_x)

# Compute length of worm from eqn (13-27) Shigley 9e, Pg. 696
L = p_x * Nw
print("Length of worm gear = %0.4f [in]" % L)

# Compute lead angle lambda from eqn (13-28) Shigley 9e, Pg. 696
lead_angle = degrees(atan(L / (pi*D_worm)))
print("Lead Angle = %0.4f [deg]" % lead_angle)


# Compute wormgear pitch diameter
p_d = 10
D_wormgear = Ng / p_d
print("Wormgear pitch diameter = %0.4f [in]" % D_wormgear)


# Compute centre distance
C = (1/2) * (D_worm + D_wormgear)
print("Centre distance C = %0.4f [in]" % C)


# Compute sliding speed Mott 6e, Pg. 456
w1 = toAngularVelocity(w_out)
r1 = iTom((1/2)*D_wormgear)
Vtg = w1 * r1 * 196.85039370079 # Convert to ft/min
print("Wormgear pitch velocity = %0.4f [ft/min]" % Vtg)

Vtw = (pi*D_worm*w_in) / 12
print("Worm pitch velocity = %0.4f [ft/min]" % Vtw)

Vgs = Vtg * (1/sin(radians(lead_angle)))
print("Sliding velocity for wormgear = %0.4f [ft/min]" % Vgs)

Vws = Vtw * (1/cos(radians(lead_angle)))
print("Sliding velocity for worm = %0.4f [ft/min]" % Vws)


# Calculate gear forces
Vs = Vws
mu = 0.103 * exp(-0.110*(Vs**0.45)) + 0.012
print("\nCoefficient of friction = %0.4f []" % mu)

Wtg = (2*torque) / D_wormgear
print("Tangential force Wtg = %0.4f [lbf]" % Wtg)

Wag = Wtg * ((cos(radians(phi_n))*sin(radians(lead_angle)) + mu*cos(radians(lead_angle))) / \
            (cos(radians(phi_n))*cos(radians(lead_angle)) - mu*sin(radians(lead_angle))))
print("Axial force Wag = %0.4f [lbf]" % Wag)

Wrg = (Wtg * sin(radians(phi_n))) / (cos(radians(phi_n)) * cos(radians(lead_angle)) - mu * sin(radians(lead_angle)))
print("Radial force Wrg = %0.4f [lbf]" % Wrg)


# Compute friction force and power
Wf = (mu*Wtg) / (cos(radians(lead_angle))*cos(radians(phi_n)) - mu*sin(radians(lead_angle)))
print("\nFriction force Wf = %0.4f [lbf]" % Wf)
power_loss = (Vs * Wf) / 33000
print("Power loss = %0.4f [hp]" % power_loss)


# Compute input power
P_out = torque * w_out / 63000
print("\nPower out = %0.4f [hp]" % P_out)
P_in = P_out + power_loss
print("Power in = %0.4f [hp]" % P_in)


# Compute efficiency
eta = P_out / P_in
print("\nEfficiency = %0.4f []" % eta)


# Gear teeth stress
Kv = 1200 * (1200 + Vtg)**(-1)
print("\nKv = %0.4f []" % Kv)
Wd = (1/Kv) * Wtg
print("Dynamic Load Wd = %0.4f []" % Wd)

y = 0.1
F = face_width = 0.625
Pn = pi * cos(radians(lead_angle/p_d))
stress = Wd / (y*F*Pn)
print("Tooth stress = %0.4f []" % stress)


# Compute rated tangential load
# Find correction factors from Mott 6e, Pg. 461
print("\nEffective face width calc = ", end="")
print(0.67*D_worm)

Cs = 1000
Dg = 4 # [in]
Fe = 0.625 # [in]
Cm = 0.815
Cv = 0.43

Wtr = Cs * Dg**0.8 * Fe * Cm * Cv
print("Wtr = %0.4f [lbf]" % Wtr)


# Compute if design can resist pitting
resistant = (Wtg < Wtr)
print("\nResistant to pitting by rated tangental force: %r" % resistant)

Dw_min = C**0.875 / 3
Dw_max = C**0.875 / 1.6

size = (D_worm > Dw_min and D_worm < Dw_max)
print("Dw_min, Dw_max = %0.4f, %0.4f [in]" % (Dw_min, Dw_max))
print("Resistant to pitting by worm permissable diameter: %r" % size)
```
